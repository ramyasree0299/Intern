**SPLITING DATA INTO TRAIN AND TEST**

```python
y = numodata
x = data1.loc[:,["age","balance"]]
from sklearn.cross_validation import train_test_split
x_train,x_test,y_train,y_test =  train_test_split(x,y,test_size = 0.2)
```

**NORMALIZATION**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_train =  scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

**USING RANDOMFOREST CLASSIFIER**

```python
from sklearn.ensemble import RandomForestClassifier
dtc_clf=RandomForestClassifier(max_depth=3,random_state=0)
dtc_clf.fit(x_train,y_train)
```

```python
y_pred=dtc_clf.predict(x_test)
y_full=dtc_clf.predict(x)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score
```

```python
dtc_cm_test=confusion_matrix(y_test,y_pred)
dtc_cm_test
plt.figure(figsize= (4,4))
sns.heatmap(dtc_cm_test,annot = True ,cbar = False)
plt.xlabel("True Values")
plt.ylabel("Predicted Values")
plt.title("Random forest model")
plt.show()
```

```python
dtc_cm_test=confusion_matrix(y_full,y)
dtc_cm_test
plt.figure(figsize= (4,4))
sns.heatmap(dtc_cm_test,annot = True ,cbar = False)
plt.xlabel("True Values")
plt.ylabel("Predicted Values")
plt.title("Random forest model")
plt.show()
```

```python
dtc_acc_test=accuracy_score(y_test,y_pred)
dtc_acc_test
```

```python
dtc_precision_test=precision_score(y_test,y_pred,average="weighted")
dtc_precision_test
```

```python
from sklearn.metrics import recall_score
dtc_recall_test=recall_score(y_test,y_pred,average="weighted")
dtc_recall_test
```
